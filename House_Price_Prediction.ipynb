```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
# Dealing with NA values
data.dropna(inplace = True)
data.info()
```

```python
from sklearn.model_selection import train_test_split
# Getting target target varible data and remaining data without it
x = data.drop(['median_house_value'], axis = 1) # Axis = 1 becuase we want to drop it along column
y = data['median_house_value']
x,y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize = (15,8))
```

```python
train_data.corr() # Now we will convert this into correlation heatmap
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(), annot = True)
#We can see that median_income correlates quite highly with median_house_value
```

```python
#Now as we see in above histograms, we have left/right skewed histogram, so we will turn them into Normal (Gaussian) distribution
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1) # we had done +1, just to be sure that we we dont have any zero value
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize = (15,8)) # Now we can observe more Guassian (Normal) distribution type nature
```

```python
train_data
# Now we also want to include the ocean_proximity feature in our model, as we could hypothyize that houses near Bay tends to receive more price. But we can't directly include them, so we will give each type of value (map) with integer values
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)
train_data
```

```python
# New heatmap with ocean_proximity features
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(), annot = True)
```

```python
# Rough map with prices
plt.figure(figsize = (15, 8))
sns.scatterplot(x = 'latitude', y = "longitude", data = train_data, hue = "median_house_value", palette = "coolwarm")
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(), annot = True)
```

```python
from sklearn.linear_model import LinearRegression

x_train, y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1) # we had done +1, just to be sure that we we dont have any zero value
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
x_test, y_test = test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
```

```python
reg.score(x_test, y_test)
```
